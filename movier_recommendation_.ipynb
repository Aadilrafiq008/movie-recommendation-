```python
# geners , id, keyword, original_language ,overview,title, cast, crew

df_preprocest = df_merge_data[['genres','movie_id', 'keywords', 'original_language', 'overview', 'title', 'cast', 'crew' ]]
```

```python
df_preprocest.isnull().sum()
df_preprocest.dropna(inplace=True)
```

```python
import ast
def sorting(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    return L
```

```python
df_preprocest['genres'] = df_preprocest['genres'].apply(sorting)
```

```python
df_preprocest['keywords'] = df_preprocest['keywords'].apply(sorting)
```

```python
def sorting_actor_names(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 4:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
df_preprocest['cast'] =  df_preprocest['cast'].apply(sorting_actor_names)

```

```python
df_preprocest['cast']
```

```python
def sorting_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
df_preprocest['crew']=  df_preprocest['crew'].apply(sorting_director)
```

```python
df_preprocest
```

```python
df_preprocest['overview'] = df_preprocest['overview'].apply(lambda x: x.split())
```

```python
df_preprocest.info()
```

```python
df_preprocest['title'].apply(lambda x: x.split())
```

```python
df_preprocest['genres'] = df_preprocest['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
df_preprocest['cast'] = df_preprocest['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
df_preprocest['crew'] = df_preprocest['crew'].apply(lambda x: [i.replace(" ", "") for i in x] )
# df_preprocest['keywords'] = df_preprocest['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
df_preprocest['tags'] = df_preprocest['genres'] + df_preprocest['overview']+ df_preprocest['cast']+ +df_preprocest['crew']
```

```python
new_df = df_preprocest[['movie_id', 'title', 'tags']] 
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
        
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features=5000, stop_words='english')

vector = CV.fit_transform(new_df['tags']).toarray()
```

```python
CV.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

```

```python
similarity = cosine_similarity(vector)
```

```python
def recommending(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        print(i[1])
```

```python
recommending('Avatar')
```
